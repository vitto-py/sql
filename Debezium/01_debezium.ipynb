# Registry API
(interface)
Reference: https://docs.confluent.io/platform/current/schema-registry/develop/api.html


* **Gives you all the schemas that are present**
```shell
    curl -XGET http://localhost:8081/schemas
```
* **All Possible schema types**
```shell
    curl -XGET http://localhost:8081/schemas/types
```

<br>

* **Lists all subjects**: here you will find your debezium connection to SQL server
```shell
    curl -XGET http://localhost:8081/subjects
```

* **All versions for a particular subject**: replace `subject` with one of the above Subjects
```shell
    curl -XGET http://localhost:8081/subjects/<subject>/versions
```
* **Schema for a particular subject and version**: you just append the number of the version at the end and you get the ACTUAL SCHEMA of that particular version
```shell
    curl -XGET http://localhost:8081/subjects/<subject>/versions/<version>
```

* **Schema for a particular subject's latest version**: Generally you need the latest one
```shell
    curl -XGET http://localhost:8081/subjects/<subject>/versions/latest
```
* **Getting a schema with id known**: Also you can access those subjects directly with the ID 
```shell
    curl -XGET http://localhost:8081/schemas/ids/<id>
```
* **Getting the schema in json format** for this particular version of this subject 
```shell
    curl -XGET http://localhost:8081/subjects/<subject>/versions/<version>/schema
```
